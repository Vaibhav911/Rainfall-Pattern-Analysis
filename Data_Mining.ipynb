```
data = pd.read_csv("Sub_Division_IMD_2017.csv")
# data = data.replace(0, np.nan) replace all 0s with Nan
data = pd.DataFrame(data)
data
```

```
months = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
subdivs = data.SUBDIVISION.unique()
print(subdivs)
```

*Median* of each region for each month. 
<br>
Here, each column refers to a region. And each row refers to a month. 
<br>
<br>
**Note**: All the means are calculated ignoring **NaN**

```
data_median = {}
for div in subdivs:
  region_median = []
  for month in months:
    region_median.append(round(data[data["SUBDIVISION"] == div][month].median(), 2)) #doing round off till 2 digits
  data_median[div] = region_median
pd.DataFrame(data_median)
```

```
data.iloc[9]["JAN": "DEC"]
count = 0
month_num = 0
for index, row in data.iterrows():
  month_num = 0
  for month in months:
    if pd.isnull(data.iloc[index][month]):
      data.at[index, month] = data_median[row['SUBDIVISION']][month_num]
    month_num += 1

# printing data with missing values filled as median
# data
```

Stratified sampling with replacement

```
strat_data = []
for div in subdivs:
  strat_data.append((data[data.SUBDIVISION == div]).sample(n=10, replace=True))
strat_data = pd.concat(strat_data)
strat_data
```

Min-Max Normalization
<br>
<br>
**Note**: Normalization done on a copy data set

```
data_norm = data.copy()
data_monthly_only = data_norm[months]
data_norm[months] = (data_monthly_only - data_monthly_only.min()) / (data_monthly_only.max() - data_monthly_only.min())
data_norm
```

State-wise aggregation of dataset


```
data_aggr = []
for div in subdivs:
  data_aggr.append(data[data['SUBDIVISION'] == div][months].mean())
data_aggr = pd.DataFrame(data_aggr)
data_aggr.insert(0, 'SUBDIVISION', subdivs)
data_aggr
```

Discretization


```
discretized_df=data.copy()
k=3
temp=0
sub_divs=data['SUBDIVISION'].unique()
#print(sub_divs)
for div in sub_divs:
  temp_df=data.loc[data['SUBDIVISION'] == div]
  for i in data.columns:
    if i=='SUBDIVISION' or i=='YEAR':
      continue
    #print(i)
    max_val=temp_df[i].max()
    min_val=temp_df[i].min()
    delta=(max_val-min_val)/3
    for j in range(temp_df.shape[0]):
      if(discretized_df.loc[j+temp,i]<=min_val+delta):
        discretized_df.loc[j+temp,i]='low'
      elif(discretized_df.loc[j+temp,i]<=min_val+2*delta):
        discretized_df.loc[j+temp,i]='medium'
      else:
        discretized_df.loc[j+temp,i]='high'
  temp+=temp_df.shape[0]
print(discretized_df)
```

Removing Noisy Data

```
no_of_intervals=10
without_noise_df=data.copy()
#print(without_noise_df)
temp=0
sub_divs=without_noise_df['SUBDIVISION'].unique()
for div in sub_divs:
  #print('div')
  temp_df=data.loc[data['SUBDIVISION'] == div]
  for i in months:
    #print(i)
    if i=='SUBDIVISION' or i=='YEAR':
      continue
    lis=temp_df[i].tolist()
    #print(lis)
    lis.sort()
    median=[]
    num=temp_df.shape[0]//no_of_intervals+1
    start=0
    end=num
    for k in range(no_of_intervals-1):
      median.append(np.median(lis[start:end+1]))
      start=end+1
      end+=num
    median.append(lis[start:])
    #print(median)
    for j in range(temp_df.shape[0]):
      temp1=lis.index(without_noise_df.loc[j+temp,i])
      without_noise_df.loc[j+temp,i]=median[temp1//(num+1)]
  temp+=temp_df.shape[0]
  #print(temp)
print(without_noise_df)
#print(data)



```
